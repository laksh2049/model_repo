```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
 dataset = pd.read_csv('Algerian_forest_fires_dataset.csv',header = 1)
```

```python
dataset.head()
```

```python
dataset.info()
```

### data clenaing 

```python
## checking for missing value 
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"] = 0
dataset.loc[122:,"Region"] = 1
df = dataset
```

```python
df.info()
```

```python
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# remove the null values 
df = df.dropna().reset_index(drop= True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# remove 122 row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix space in column names 
df.columns= df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
### change the required colum as integer data type
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
### changinf the othercolumns to float datatype 
objects = [features for features in df.columns if df[features].dtypes =='O']
```

```python
for i in objects :
    if i!= 'Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
##lest save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index= False)
```

```python
### drop day month year
df_copy = df.drop(['day','month','year'],axis = 1)
```

```python
df_copy['Classes'] = df_copy['Classes'].astype('string')
```

```python
### visulaization (density plot for all the feature)
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()  # Applies seaborn styling
df_copy.hist(bins=50, figsize=(20, 15))
plt.show()


```

```python
## percantage for pie chart 
percentage=df_copy['Classes'].value_counts(normalize = True)*100
```

```python
## plotting pie chart 
classlables = ["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels = classlabels,autopct='%1.1f%%')
plt.title("pie chart of classes")
plt.show()

```
